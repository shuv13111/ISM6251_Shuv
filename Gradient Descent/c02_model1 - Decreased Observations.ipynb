# Predictive Modeling Example

## Step 1: Import the libraries we will use in this notebook

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

np.random.seed(1) # set this to ensure the results are repeatable. 
```

## Step 2: Load the given data that we will model

```python
df = pd.read_csv('c02_dataset_1_v1_decreased_observation.csv')

df.head(3)
```

Seperate out the input m(X) and the target (y)

```python
X=df[['input']]
y=df[['target']]
```

For educational purposes, let's look at the first five values in X and y (note how X is an array of lists)

## Step 3: Explore the given data

Often we will do much more than this, but at least look at things using a scatterplot

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.scatter(X, y, color='red')

ax.set_xlabel('input')
ax.set_ylabel('target')
plt.tight_layout()
plt.show()
```

## Step 4: Fit the model

Fit the linear regression model to the dataset

```python
lin_reg=LinearRegression()
_ = lin_reg.fit(X,y)

# OR, we could combine the two lines above to one as follows...
#lin_reg = LinearRegression().fit(X,y)
```

## Step 5: Analyze the linear regression model's performance

Visualize the linear regression model results

```python
fig = plt.figure()
ax = fig.add_subplot()

ax.scatter(X, y, color='red')
ax.scatter(X, lin_reg.predict(X), color='blue')

ax.set_title('Linear Regression')
ax.set_xlabel('input')
ax.set_ylabel('target')
plt.tight_layout()
plt.show()
```

```python
b0 = lin_reg.intercept_[0]
b1 = lin_reg.coef_[0][0]
r2 = lin_reg.score(X, y)
```

... using fstring formatting, we can display these values as follows...

```python
print(f"Y = {b0:.2f} + {b1:.2f}x")
print(f"R^2: {lin_reg.score(X, y):.3f}") 
# for more on fstrings see here...
# https://www.freecodecamp.org/news/python-f-strings-tutorial-how-to-use-f-strings-for-string-formatting/
```

## Interpretation of the model compared to the original model

Output of model developed during the class:

Y = -0.33 + 30.73x | R^2: 0.896

Output of the model with reduced observations:

Y = 0.43 + 30.72x | R^2: 0.907

Upon comparing the two models, it can be observed that the reduction in observations did not result in a decrease in the R-squared value. Instead, it slightly increased from 0.896 to 0.907. However, there was a noticeable difference in the intercept values of the two models, where the first model had an intercept of -0.33 and the second model had an intercept of 0.43. This difference may indicate that the two models were fitted to different subsets of data, which can influence the estimated intercept and slope coefficients.

Specifically, the reduced observation is the result of decreasng the sample size from 1000 to 500.
